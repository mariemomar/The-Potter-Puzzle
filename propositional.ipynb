```python
from aima.logic import *

# Knowledge base
kb = PropKB()

# The propositions (now as `expr`)
MY_MON = expr("Mummy_Yoga_Monday")
MY_THU = expr("Mummy_Yoga_Thursday")
IC = expr("Love_IceCream")   # ic --> dp
DP = expr("Dreams_Of_Paris")
LC = expr("Loving_Cream")
BM = expr("Betty_Likes_Marmalade") # only
MM = expr("Mummy_Eats_Marshmallows") # only
# three boxes
TS = expr("Trip_to_Sea")
SL = expr("Swan_Lake_Ticket")
NAC = expr("New_Album_Coins")

ASP = expr("Aunt_Polly_Sewing") # and materials
BS = expr("Betty_Ballet_Suit") # for classes
PF = expr("Peter_Goes_Fishing")
PRC = expr("Peter_Collects_Rare_Coins") # new album coins 
PNC = expr("Peter_No_Cream") # no cake

SAME_WISH = expr("Mummy_and_Daddy_Same_Wish") 

NAPOLEON_CAKE = expr("Favourite_Dessert_Napoleon_Cake")
MARMALADE = expr("Favourite_Dessert_Marmalade")
WAFFLES = expr("Favourite_Dessert_Waffles")



# Define relationships with individuals
MUMMY_LIKES_NAP_CAKE = expr("Mummy_Likes_Napoleon_Cake")
MUMMY_DREAMS_PARIS = expr("Mummy_Dreams_Paris")

BETTY_LIKES_NAP_CAKE = expr("Betty_Likes_Napoleon_Cake")
BETTY_DREAMS_PARIS = expr("Betty_Dreams_Paris")

PETER_DREAMS_PARIS = expr("Peter_Dreams_Paris")
PETER_LIKES_NAP_CAKE = expr("Peter_Likes_Napoleon_Cake")

AUNTPOLLY_DREAMS_PARIS = expr("AuntPolly_Dreams_Paris")
AUNTPOLLY_LIKES_NAP_CAKE = expr("AUNTPOLLY_Likes_Napoleon_Cake")

DADDY_GOES_FISHING = expr("Daddy_Goes_Fishing")
DADDY_WHISHING_GOING_TO_SEA = expr("Daddy_Goes_To_Sea")
DADDY_LIKES_NAP_CAKE = expr("Daddy_Likes_Napoleon_Cake")
DADDY_DREAMS_PARIS = expr("Daddy_Dreams_Paris")
kb.tell(DADDY_GOES_FISHING)


# Facts
kb.tell(MY_MON)
kb.tell(MY_THU)
kb.tell(IC)
kb.tell(DP)
kb.tell(BM)
kb.tell(MM)

kb.tell(TS)
kb.tell(SL)
kb.tell(NAC)

kb.tell(ASP)
kb.tell(BS)
kb.tell(PNC)
kb.tell(SAME_WISH)
kb.tell(NAPOLEON_CAKE)
kb.tell(MARMALADE)
kb.tell(WAFFLES)
kb.tell(PF)
kb.tell(PRC)

# daddy goes fishing then its wish is to go to the sea , and mummy has the same wish as daddy


kb.tell(expr("Peter_Goes_Fishing ==> Peter_Collects_Rare_Coins"))
kb.tell(expr("Peter_No_Cream ==> ~ Peter_Dreams_Paris")) # true
kb.tell(expr("Mummy_Eats_Marshmallows ==> ~ Mummy_Likes_Napoleon_Cake")) # true
kb.tell(expr("Betty_Likes_Marmalade ==> ~Betty_Likes_Napoleon_Cake")) # true
kb.tell(expr("Daddy_Goes_Fishing ==> Daddy_Goes_To_Sea"))
kb.tell(expr("Daddy_Goes_To_Sea ==> ~ Daddy_Dreams_Paris"))
kb.tell(expr("Daddy_Goes_To_Sea ==> ~ Mummy_Dreams_Paris"))
kb.tell(expr("Betty_Likes_Marmalade ==> ~Betty_Dreams_Paris"))
kb.tell(expr("Peter_No_Cream ==> ~ Peter_Likes_Napoleon_Cake"))


# Query to determine "Who dreams of Paris?"
def who_dreams_of_paris():
    individuals = [
        ("Mummy", MUMMY_DREAMS_PARIS),
        ("Peter", PETER_DREAMS_PARIS),
        ("AuntPolly", AUNTPOLLY_DREAMS_PARIS),
        ("Betty", BETTY_DREAMS_PARIS),
        ("Daddy", DADDY_DREAMS_PARIS)
    ]
    return [
        name
        for name, prop in individuals
        if dpll_satisfiable(expr(f"{prop} & {' & '.join(map(str, kb.clauses))}"))
    ]

# Query to determine "Who likes Napoleon cake?"
def who_likes_napoleon_cake():
    individuals = [
        ("Mummy", MUMMY_LIKES_NAP_CAKE),
        ("Betty", BETTY_LIKES_NAP_CAKE),
        ("Peter" , PETER_LIKES_NAP_CAKE),
        ("AuntPolly", AUNTPOLLY_LIKES_NAP_CAKE),
        ("Daddy" ,  DADDY_LIKES_NAP_CAKE)
    ]
    return [
        name
        for name, prop in individuals
        if dpll_satisfiable(expr(f"{prop} & {' & '.join(map(str, kb.clauses))}"))
    ]



# Results
print("Who dreams of Paris?")
print(who_dreams_of_paris())

print("Who likes Napoleon cake?")
print(who_likes_napoleon_cake())

```
